```python
# Generar un CRUD (Crear, Leer, Actualizar y Eliminar) de una agenda de telefono.
import os
import time

class Contacto(object):
    def __init__(self, nombre, telefono):
        self.nombre=nombre
        self.telefono=telefono
        
    def leer(self):
        l = [self.nombre,self.telefono]
        #return "Nombre: "+str(self.nombre) + "  Telefono: " +str(self.telefono)
        return l
    
    def setNombre(self, nom):
        print(nom)
        self.nombre = nom
        return self.nombre
        
    def setTelefono(self, telf):
        print(telf)
        self.telefono = telf
        return self.telefono

contactos = []
banSalir = False

#Lista precargada
contactos.append(Contacto("roberto","11"))
contactos.append(Contacto("carlos","22"))
contactos.append(Contacto("mateo","33"))
contactos.append(Contacto("sebastian","44"))

def clear():
    os.system("cls")

def creaContacto():
    clear()
    print("INGRESAR CONTACTO")
    auxNombre = input("Ingrese el nombre de contacto: ")
    auxTelefono = input("Ingrese el telefono de contacto: ")
    contactos.append(Contacto(auxNombre,auxTelefono))
    print("Contacto agregado exitosamente")
    print(contactos[len(contactos)-1].leer())


def busca(crit):
    #genera una lista de contactos
    auxL = []
    auxE = -1
    if len(contactos)>0:
        for i in range(0,len(contactos)):           
            auxL.append(contactos[i].leer())
            if crit in auxL[i]:
                auxE=i
    return auxE

    
def buscaContacto():
    clear()
    print("BUSCAR CONTACTO")
    crit = input("Ingrese el criterio de busqueda: ")    
    auxE = busca(crit) 
    if auxE != -1:
            print("Coincidencia Encontrada: ")
            l = contactos[auxE].leer()
            print("Nombre: " , l[0])
            print("Telefono: " , l[1])
            return auxE
    else:
        print("No existen contactos")
        return -1

def modContacto():
    clear()
    print("MODIFICAR CONTACTO")
    crit = input("Ingrese el criterio de busqueda: ")    
    auxE = busca(crit) 
    if auxE != -1:
            print("Coincidencia Encontrada: ")
            l = contactos[auxE].leer()
            print("Nombre: " , l[0])
            print("Telefono: " , l[1])
            n = input("Ingrese el nuevo nombre o deje en blanco para no cambiar: ")
            t = input("Ingrese el nuevo telefono o deje en blanco para no cambiar: ")
            
            if n!="":
                print("Nombre cambiado a : ", contactos[auxE].setNombre(n))
            
            if t!="":
                print("Telefono cambiado a: ", contactos[auxE].setTelefono(t))
    else:
        print("No existe el contacto")

def eliminaContacto():
    clear()
    print("ELIMINAR CONTACTO")
    crit = input("Ingrese el criterio de busqueda: ")    
    auxE = busca(crit) 
    if auxE != -1:
            print("Coincidencia Encontrada: ")
            l = contactos[auxE].leer()
            print("Nombre: " , l[0])
            print("Telefono: " , l[1])
            r = input("Presione s si desea eliminar el contacto : ")
            if r== "s":
                contactos.pop(auxE)
                print("Contacto eliminado")
    
def printMenu():
    clear()
    print()
    print("DIRECTORIO TELEFONICO")
    print("1 Ingresar contacto")
    print("2 Consultar contacto")
    print("3 Modificar contacto")
    print("4 Eliminar contacto")
    print("5 Salir")

def seleccionaMenu():
    printMenu()
    ban = True
    while ban==True:
        op = int(input("Seleccione una opcion: "))
        if op >0 and op<6:
            ban=False
            return op
        

while banSalir==False:
    op=seleccionaMenu()
    if op==1:
        #Ingresar nuevo contacto
        creaContacto()
    elif op==2:
        buscaContacto()
    elif op==3:
        modContacto()
    elif op==4:
        eliminaContacto()
    elif op==5:
        print("Programa finalizado")
        banSalir=True
    
```

```python
# Crear un metodo de validacion de cedula Ecuatoriana, en caso de que la cedula no sea validad lanzar 
# una excepcion, ademas de controlar que solo pueda ingresar digitos numericos por teclado.

def validaCed(ced):
    print("Validando: " , ced)
    cp = int(ced[0:2])
    if cp >= 1 and cp <= 24: # verificar codigo de provincia
        tercer_dig = int(ced[2])
        if tercer_dig >= 0 and tercer_dig < 6 : # numeros enter 0 y 5
            return __validar_ced(ced)
        else:
            raise Exception(u'Tercer digito invalido') 
    else:
        raise Exception(u'Codigo de provincia incorrecto')

def __validar_ced(nro):
    total = 0
    base = 10
    d_ver = int(nro[9])# digito verificador
    multip = (2, 1, 2, 1, 2, 1, 2, 1, 2)
    for i in range(0,len(multip)):
        p = int(nro[i]) * multip[i]
        total+=p if p < 10 else int(str(p)[0])+int(str(p)[1])      
    mod = total % base
    val = base - mod if mod != 0 else 0
    return val == d_ver
    
    
ban=True

while ban == True:
    ingreso = input("Ingresa un numero de cedula: ")
    #Verifica que se ingreso solo numeros
    try:
        auxIngreso=int(ingreso)
    except ValueError as e:
        print("El numero de cedula no puede contener letras o estar vacio")
        #print(e)
    else:
        # Si no existe ninguna excepcion no ocurre nada.
        if len(ingreso)==10:
            try:
                print("validacion = " , validaCed(ingreso))
            except ValueError as e:
                print(e)
            finally:
                ban=False
        else:
            print("La cedula debe tener 10 digitos")
        pass
```
